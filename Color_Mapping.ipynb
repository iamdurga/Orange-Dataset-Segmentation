```python
plt.imshow(img)
```

## Converting to RGB ColorSpace

```python
img_fix = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_fix)
```

```python
img_fix.shape
```

Our color image have 4160 rows and 1872 columns.

## Converting to HSV

```python
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
plt.imshow(img_hsv)
```

## Lets detect the orange from image

To detect the only orange from our image, we use array slicing while observing our orange in the picture.

```python
only_orange = img_fix[800: 1700, 350:1360]
plt.imshow(only_orange)
```

## Inverse of the image

`bitwise_not()` filp the pixels of the image that mean it change color from white to black vice versa.

```python
img_inv = cv2.bitwise_not(img_fix)
plt.imshow(img_inv)
```
